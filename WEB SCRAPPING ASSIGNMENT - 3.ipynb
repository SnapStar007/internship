```python
#code to scrape product delivery time and date
delivery=[]
for i in page:
    driver.get(i)
    get=driver.find_elements_by_xpath("//div[@class='a-row s-align-children-center']")
    try:
        for j in get:
            delivery.append(j.text)
    except:
        delivery.append('-')
delivery
```

# 5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
driver=webdriver.Chrome('chromedriver.exe')
url="https://www.google.co.in/maps"
driver.get(url)
```

```python
city = input('Enter City Name : ')                                       
search = driver.find_element_by_id("searchboxinput")                    
search.clear()                                                             
time.sleep(2)
search.send_keys(city)                                                    
button = driver.find_element_by_id("searchbox-searchbutton")              
button.click()                                                          
time.sleep(3)

try:
    url_string = driver.current_url
    print("URL Extracted: ", url_string)
    lat_lng = re.findall(r'@(.*)data',url_string)
    if len(lat_lng):
        lat_lng_list = lat_lng[0].split(",")
        if len(lat_lng_list)>=2:
            lat = lat_lng_list[0]
            lng = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(lat, lng))

except Exception as e:
        print("Error: ", str(e))
```

# 6. Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 – September 20) from trak.in.

```python
driver=webdriver.Chrome('chromedriver.exe')
url='https://trak.in/'
driver.get(url)
```

```python
#code to click on funding deals in trak.in
fund=driver.find_element_by_xpath("//a[@title='http://trak.in/india-startup-funding-investment-2015/']")
fund.click()
```

```python
#code to scape all the column details in 2nd quarter
Date=[]
Startup=[]
Industry=[]
SubVertical=[]
Location=[]
Investor=[]
Investment=[]
Amount=[]

for i in range(48,51):
    dt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[2]'.format(i))
    for d in dt:
        Date.append(d.text)
        
    st=driver.find_elements_by_xpath("//table[@id='tablepress-{}']/tbody/tr/td[3]".format(i))
    for s in st:
        Startup.append(s.text)
        
    ind=driver.find_elements_by_xpath("//table[@id='tablepress-{}']/tbody/tr/td[4]".format(i))
    for n in ind:
        Industry.append(n.text)
    
    sv=driver.find_elements_by_xpath("//table[@id='tablepress-{}']/tbody/tr/td[5]".format(i))
    for v in sv:
        SubVertical.append(v.text)
        
    loc=driver.find_elements_by_xpath("//table[@id='tablepress-{}']/tbody/tr/td[6]".format(i))
    for l in loc:
        Location.append(l.text)
    
    inv=driver.find_elements_by_xpath("//table[@id='tablepress-{}']/tbody/tr/td[7]".format(i))
    for u in inv:
        Investor.append(u.text)
    
    it=driver.find_elements_by_xpath("//table[@id='tablepress-{}']/tbody/tr/td[8]".format(i))
    for t in it:
        Investment.append(t.text)
    
    amt=driver.find_elements_by_xpath("//table[@id='tablepress-{}']/tbody/tr/td[7]".format(i))
    for a in amt:
        Amount.append(a.text)
```

```python
#saving the scraped details in a Dataframe
df=pd.DataFrame()
df['Date']=Date
df['Startup']=Startup 
df['Industry']=Industry
df['SubVertical']=SubVertical
df['Location']=Location
df['Investor']=Investor
df['Investment']=Investment
df['Amount']=Amount
df
```

# 10. Write a python program to scrape a data for all available Hostels from 
https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, 
distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms 
from price, facilities and property description.

```python
driver=webdriver.Chrome('chromedriver.exe')
url='https://www.hostelworld.com/'
driver.get(url)
```

```python
#search london in location bar
loca=driver.find_element_by_xpath("//input[@id='search-input-field']")
loca.send_keys('london')
#select london
london = driver.find_element_by_xpath('/html/body/div[1]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div')
```

```python
#click the search bar
search=driver.find_element_by_xpath("//button[@id='search-button']")
search.click()
```

```python
#creating empty list to store all the data
hostel= []
distance = []
pvt_prices = []
dorms_price = []
rating = []
reviews = []
over_all = []
facilities = []
description =[]
product_url = []
```

```python
#scrapping hostel name
name = driver.find_elements_by_xpath("//h2[@class='title title-6']")
for i in name:
    hostel.append(i.text)
```

```python
#scraping distance from city center
dist = driver.find_elements_by_xpath("//div[@class='subtitle body-3']//a//span[1]")
for i in dist:
    distance.append(i.text)
```

```python
#scraping privates price
pvt_price = driver.find_elements_by_xpath("//div[@class='prices-col']/a/div[1]")
for i in pvt_price:
    try:
        pvt_prices.append(i.text.replace('\n',' '))
    except:
        pvt_prices.append('-')
pvt_prices
```

```python
#scraping dorms price
dorms = driver.find_elements_by_xpath("//div[@class='prices-col']/a/div[2]")
for i in dorms:
    dorms_price.append(i.text.replace('\n',' '))
dorms_price
```

```python
#scrapping facilities 
facil=driver.find_elements_by_xpath("//div[@class='facilities-label facilities']")
for i in facil:
    facilities.append(i.text.replace('\n'," "))
facilities
```

```python
url=driver.find_elements_by_xpath("//div[@class='prices-col']//a[2]")
for i in url:
    product_url.append(i.get_attribute('href'))
for i in product_url:
    driver.get(i)
    time.sleep(1)
    #lets click on show more button for description
    try:
        driver.find_element_by_xpath("//a[@class='toggle-content']").click()
        time.sleep(1)
    except NoSuchElementException:
        pass
    #scraping ratings
    try:
        rate=driver.find_element_by_xpath("//div[@class='score orange big' or @class='score gray big']")
        rating.append(rate.text)
    except:
        rating.append('-')
    #scraping total reviews
        
    try:
        revi = driver.find_element_by_xpath("//div[@class='reviews']")
        reviews.append(revi.text.replace('Total Reviews',''))
    except NoSuchElementException:
        reviews.append('-')
        #scraping overall review
    try:
        overall=driver.find_element_by_xpath("//div[@class='keyword']//span")
        over_all.append(overall.text)
    except:
        over_all.append('-')
    #scraping property description 
    try:
        disc = driver.find_element_by_xpath("//div[@class='content']")
        description.append(disc.text)
    except:
        description.append('-')
```

```python
#storing all the data in a DataFrame
df = pd.DataFrame()
df['Hostel_Name']=hostel[0:30]
df['Distance fron city centre']=distance[0:30]
df['Ratings']=rating[0:30]
df['Total_reviews']=reviews[0:30]
df['Overall Reviews']=over_all[0:30]
df['Privates from price']=pvt_prices[0:30]
df['Dorms from price']=dorms_price[0:30]
df['Property Description']=description[0:30]
df
```
