# 2) Scrape the details team India’s international fixtures from bcci.tv.

```python
#to get to url page
driver=webdriver.Chrome('chromedriver.exe')
url='https://www.bcci.tv/'
driver.get(url)
```

```python
#to access the international fixtures fo india
btn1=driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover'][1]")
btn1.click()
btn2=driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down'][1]")
btn2.click()
```

```python
#to scrape match title
title=driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")

match=[]
try:
    for i in title:
        match.append(i.text)
except NoSuchElementException:
    match.append('N/A')
match
```

```python
#to scrape the series of the match
ser=driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")

series=[]
try:
    for i in ser:
        series.append(i.text)
except NoSuchElementException:
    series.append('N/A')
series
```

```python
#to scrape details of the place
pla=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span")

place=[]
try:
    for i in pla:
        place.append(i.text)
except NoSuchElementException:
    place.append('N/A')
place
```

```python
urls=[]
page=driver.find_elements_by_xpath("//div[@class='js-list']/a")
for i in page:
    urls.append(i.get_attribute('href'))
```

```python
date=[]
time=[]
for i in urls:
    driver.get(i)
    
    try:
        dat = driver.find_element_by_xpath("//div[@class='mc-header-scorebox__datetime']/strong")
        date.append(dat.text)
    except NoSuchElementException:
        date.append('N/A')
        
    try:
        times = driver.find_element_by_xpath("//div[@class='mc-header-scorebox__datetime']/span")
        time.append(times.text)
    except NoSuchElementException:
        time.append('N/A')
```

```python
df=pd.DataFrame({"Match":match,
                "Series":series,
                "Place":place,
                "Date":date,
                "Time":time})
df
```

# 5) Trending repositories from github.com

```python
#to get to url page
driver=webdriver.Chrome('chromedriver.exe')
url='https://github.com'
driver.get(url)
```

```python
#to get the dropdown menu of explore
explore=driver.find_element_by_xpath("//nav[@class='mt-0 px-3 px-lg-0 mb-5 mb-lg-0']/ul/li[4]")
explore.click()
```

```python
#to click on trending option from explore
trending=driver.find_element_by_xpath("//div[@class='dropdown-menu flex-auto rounded px-0 pt-2 pb-0 mt-0 pb-4 p-lg-4 position-relative position-lg-absolute left-0 left-lg-n4']/ul[2]/li[3]/a")
trending.click()
```

```python
#to scrape the details of the title or name of the repository
name=[]
try:
    names=driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/h1/a")
    for i in names:
        name.append(i.text)
except NoSuchElementException:
    name.append('N/A')
name
```

```python
#to scrape the description
desc=[]
try:
    de=driver.find_elements_by_xpath("//article[@class='Box-row']/p")
    for i in de:
        desc.append(i.text)
except NoSuchElementException:
    desc.append('N/A')
desc
```

```python
#to scrape url of the repository 
urls=[]
repo=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in repo:
    urls.append(i.get_attribute('href'))
```

```python
#to scrape the contributors
Contributors=[]
Language=[]
for i in urls:
    driver.get(i)
    l=[]
    try:
        count=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        Contributors.append(count.text)
    except NoSuchElementException:
        Contributors.append('No details available')
```

```python
#to scrape language 
language=[]
try:
    for i in urls:
        driver.get(i)
        count =driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]')    
        listy=count.text.split('\n')
        if 'Languages' in listy:
            language.append(listy[listy.index('Languages')+1])
        else:
            language.append('N/A')
except NoSuchElementException:
    language.append('N/A')
```

```python
df=pd.DataFrame()
df['Repository title']=name[:24]
df['Repository description']=desc[:24]
df['Language used']=language[:24]
df['Contributors count']=Contributors[:24]
df.style.hide_index()
```
